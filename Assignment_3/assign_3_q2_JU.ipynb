```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.datasets import cifar10
```

# MNIST Dataset

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train
```

```python
y_train
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
```

```python
x_train, x_test =  x_train/255.0, x_test/255.0
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))


model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

model.summary()

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Reshape data to include channel dimension
# x_train = x_train.reshape(-1, 28, 28, 1)
# x_test = x_test.reshape(-1, 28, 28, 1)

history = model.fit(x_train, y_train, epochs=10,
                    validation_data=(x_test, y_test))

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print(test_acc * 100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

cm = confusion_matrix(y_test, y_pred_classes)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.arange(10), yticklabels=np.arange(10))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix Heatmap')
plt.show()

y_test_bin = label_binarize(y_test, classes=np.arange(10))

fpr, tpr, roc_auc = {}, {}, {}
for i in range(10):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 8))
for i in range(10):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

fpr_micro, tpr_micro, _ = roc_curve(y_test_bin.ravel(), y_pred.ravel())
roc_auc_micro = auc(fpr_micro, tpr_micro)

print(f'Micro-average AUC: {roc_auc_micro:.2f}')
```

# CIFER-10 Dataset

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import datasets, layers, models
```

```python
x_train, x_test =  x_train/255.0, x_test/255.0
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))


model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))

model.summary()

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(x_train, y_train, epochs=10,
                    validation_data=(x_test, y_test))

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)
print(test_acc * 100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

cm = confusion_matrix(y_test, y_pred_classes)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.arange(10), yticklabels=np.arange(10))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix Heatmap')
plt.show()

y_test_bin = label_binarize(y_test, classes=np.arange(10))

fpr, tpr, roc_auc = {}, {}, {}
for i in range(10):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 8))
for i in range(10):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

fpr_micro, tpr_micro, _ = roc_curve(y_test_bin.ravel(), y_pred.ravel())
roc_auc_micro = auc(fpr_micro, tpr_micro)

print(f'Micro-average AUC: {roc_auc_micro:.2f}')
```

# Discussion

Here is an elaborated discussion of the model's performance on the MNIST and CIFAR-10 datasets based on the results from the notebook:

**MNIST Dataset:**

*   **Accuracy:** The model achieved a high test accuracy of **98.87%** on the MNIST dataset. This indicates that the model is very effective at classifying handwritten digits.
*   **Loss:** The test loss was **0.0490**. The low loss value, coupled with high accuracy, further confirms the model's strong performance.
*   **Training Performance:** Looking at the training history plot, the training accuracy rapidly increased and the training loss decreased over the epochs. The validation accuracy also improved significantly, reaching a high value, indicating that the model generalized well to unseen data.
*   **Confusion Matrix:** The confusion matrix showed very few misclassifications, with high numbers along the diagonal, indicating correct predictions for most digits.
*   **ROC Curve and AUC:** The ROC curves for individual classes and the micro-average AUC of **1.00** demonstrate excellent discriminative power of the model for separating the different digit classes.

**CIFAR-10 Dataset:**

*   **Accuracy:** The model achieved a test accuracy of **70.25%** on the CIFAR-10 dataset. This is considerably lower than the accuracy on MNIST.
*   **Loss:** The test loss was **0.8780**. This higher loss compared to MNIST reflects the increased difficulty of the CIFAR-10 classification task.
*   **Training Performance:** The training accuracy and validation accuracy increased over epochs, and the loss decreased, but the overall accuracy remained lower than on MNIST. There is a noticeable gap between training and validation accuracy towards the later epochs, which might suggest some overfitting, although not severe.
*   **Confusion Matrix:** The confusion matrix for CIFAR-10 shows more off-diagonal values compared to MNIST, indicating more misclassifications and confusions between different classes. This is expected due to the visual similarities between some CIFAR-10 classes.
*   **ROC Curve and AUC:** The micro-average AUC of **0.93** is still quite good, but lower than MNIST. The individual class ROC curves and AUC values show more variation, highlighting the difference in the model's ability to classify certain classes.

**Reasons for the Variation in Results:**

The significant difference in performance between the two datasets is primarily due to their inherent characteristics:

*   **Complexity:** CIFAR-10 images are more complex than MNIST images. They are in color (3 channels vs 1 channel), have higher resolution (32x32 vs 28x28), and contain more varied backgrounds and textures.
*   **Variability:** The classes in CIFAR-10 have much higher within-class variability (e.g., different breeds of dogs, different types of vehicles) compared to the relatively consistent style of handwritten digits in MNIST.
*   **Features:** Recognizing objects in natural images (CIFAR-10) requires learning more complex and abstract features than recognizing simple digit shapes (MNIST).

In summary, the model, which performed exceptionally well on the simpler MNIST dataset, shows a reasonable but lower performance on the more challenging CIFAR-10 dataset. This highlights the impact of dataset complexity and variability on the performance of a convolutional neural network.
